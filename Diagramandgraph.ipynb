```python
import pandas as pd
import numpy as np
from chembl_webresource_client.new_client import new_client
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv('/content/drive/MyDrive/Susan_project/All_chembl_id (1).csv')
df_1
```

```python
df_1.info()
```

```python
df_1 = df_1.drop_duplicates()
```

```python
# i dropped the columns with value NaN
drop_columns = ['Unnamed: 0','assay_variant_accession','assay_variant_mutation','text_value','toid','upper_value', 'action_type', 'activity_comment', 'data_validity_comment', 'data_validity_description','ligand_efficiency','pchembl_value','standard_text_value','standard_upper_value', 'activity_properties']
df_1 = df_1.drop(columns=drop_columns)
df_1
```

```python
# i dropped the columns with value NaN построить график где видно все организмы, уникальные организмы
drop_columns = ['qudt_units', 'assay_chembl_id', 'document_chembl_id','parent_molecule_chembl_id']
df_1 = df_1.drop(columns=drop_columns)
# Deleting rows where the value in the target_organism column is 'Homo sapiens'
df_1 = df_1[(df_1['target_organism'] != 'Homo sapiens') &
                         (df_1['target_organism'] != 'Mus musculus')]
df_1.reset_index(drop=True, inplace=True)
display(df_1)
```

```python
df_1.dropna(subset=['target_organism'], inplace=True)
df_1
```

```python
df_1.info()
```

```python
# Сравнение значений в двух столбцах и получение индексов, где они разные
different_indices = df_1.index[df_1['standard_value'] != df_1['value']].tolist()

# Вывод индексов
print("Индексы строк с разными значениями:", different_indices)
```

```python
df_1["type"].value_counts().plot.pie()
# Нужно оставить индикаторы которые отностяся к коенцентрации и убрать все остальные
```

```python
# Подсчитываем уникальные значения в столбце 'category'
value_counts = df_1['type'].value_counts()
# Выводим количество уникальных значений
display(value_counts)
```

Removing the values less then 100 count

```python
# Создаем маску для значений, которые встречаются 100 раз или более
mask = df_1['type'].isin(value_counts[value_counts >= 100].index)

# Фильтруем DataFrame, оставляя только те строки, которые соответствуют маске
filtered_df = df_1[mask]

# Выводим результат
display(filtered_df)
```

```python
# Построение столбчатой диаграммы частоты
plt.figure(figsize=(8, 6))  # Устанавливаем размер фигуры
value_counts.plot(kind='bar', color='skyblue')  # Столбчатая диаграмма
plt.title('Frequency of Categories')  # Заголовок
plt.xlabel('Categories')  # Метка оси X
plt.ylabel('Frequency')  # Метка оси Y
plt.xlim(0, 300)  # Установите пределы по оси X
plt.ylim(0, 200)
plt.xticks(rotation=45)  # Поворачиваем метки оси X для удобства чтения
plt.show()  # Показываем график
```
