```python
import pandas as pd
from chembl_webresource_client.new_client import new_client

# Загрузим исходную базу данных
data = df

# Проверим первые несколько строк данных
print(data.head())

# Инициализация клиента ChEMBL
molecule = new_client.molecule
activity = new_client.activity

# Функция для получения данных из ChEMBL по названию лекарства
def get_chembl_data(drug_name):
    try:
        # Поиск молекулы по названию
        molecule_data = molecule.search(drug_name)
        if molecule_data:
            chembl_id = molecule_data[0]['molecule_chembl_id']
            smiles = molecule_data[0]['molecule_structures']['canonical_smiles'] if 'molecule_structures' in molecule_data[0] else None
            
            # Поиск активности по ChEMBL ID
            activities = activity.filter(molecule_chembl_id=chembl_id).only(['assay_chembl_id', 'activity_comment', 'activity_id', 'assay_description', 'standard_type', 'standard_value', 'target_chembl_id', 'target_organism'])
            activity_list = []
            
            for act in activities:
                standard_type = act.get('standard_type')
                standard_value = act.get('standard_value')
                target_organism = act.get('target_organism')
                if standard_type and standard_value and target_organism:
                    activity_list.append({
                        'Name': drug_name,
                        'ChEMBL ID': chembl_id,
                        'SMILES': smiles,
                        'Activity Type': standard_type,
                        'Activity Value': standard_value,
                        'Target Organism': target_organism
                    })
            
            return activity_list
        else:
            return [{'Name': drug_name, 'ChEMBL ID': None, 'SMILES': None, 'Activity Type': None, 'Activity Value': None, 'Target Organism': None}]
    except Exception as e:
        print(f"Error fetching data for {drug_name}: {e}")
        return [{'Name': drug_name, 'ChEMBL ID': None, 'SMILES': None, 'Activity Type': None, 'Activity Value': None, 'Target Organism': None}]

# Применяем функцию к каждому лекарству и собираем все данные в один список
all_activities = []
for drug in data['Name']:
    all_activities.extend(get_chembl_data(drug))

# Создаем DataFrame из списка всех активностей
activities_df = pd.DataFrame(all_activities)

# Сохраняем новую базу данных в CSV файл
output_file = 'path_to_your_output_file.csv'  # Замените на путь к вашему файлу
# activities_df.to_csv(output_file, index=False)

print("Данные успешно обновлены и сохранены в", output_file)

```
